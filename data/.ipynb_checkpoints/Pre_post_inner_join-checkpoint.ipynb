```python
import pandas as pd
```

```python
pre = pd.read_csv("2021_Alumnos_pre_normalizadas.csv")
```

```python
post = pd.read_csv("2021_Alumnos_post_normalizadas.csv")
```

```python
pre = pre.drop_duplicates(subset = 'Codigo', keep = 'first')
```

```python
post = post.drop_duplicates(subset = 'Codigo', keep = 'last')
```

```python
pre_post = pre.merge(post, on = 'Codigo', suffixes=('_PRE', '_POST'))
pre_post
```

```python
sum(pre_post.duplicated(subset='Codigo'))
```

```python
pre_post.to_csv("2021_pre_post_normalizadas.csv")
```
